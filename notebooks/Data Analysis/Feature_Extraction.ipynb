```python
import pandas as pd

# Assuming you have your sample DataFrame 'sample_df'

# Create a function to calculate statistics
def calculate_statistics(data):
    min_value = data.min()
    max_value = data.max()
    mean_value = data.mean()
    std_value = data.std()
    return min_value, max_value, mean_value, std_value

# Group by 'gene_id' and 'label' and calculate statistics
grouped = sample_df.groupby(['gene_id', 'label'])

# Create an empty DataFrame to store the results
results = pd.DataFrame(columns=['gene_id', 'label', 'min_left_dwell', 'max_left_dwell', 'mean_left_dwell', 'std_left_dwell',
                               'min_mid_dwell', 'max_mid_dwell', 'mean_mid_dwell', 'std_mid_dwell',
                               'min_right_dwell', 'max_right_dwell', 'mean_right_dwell', 'std_right_dwell'])

# Loop through the groups and calculate statistics
for (gene_id, label), group in grouped:
    min_left_dwell, max_left_dwell, mean_left_dwell, std_left_dwell = calculate_statistics(group['left_dwell'])
    min_mid_dwell, max_mid_dwell, mean_mid_dwell, std_mid_dwell = calculate_statistics(group['mid_dwell'])
    min_right_dwell, max_right_dwell, mean_right_dwell, std_right_dwell = calculate_statistics(group['right_dwell'])
    
    # Append the results as a new row to the 'results' DataFrame
    results = pd.concat([results, pd.DataFrame({
        'gene_id': [gene_id],
        'label': [label],
        'min_left_dwell': [min_left_dwell],
        'max_left_dwell': [max_left_dwell],
        'mean_left_dwell': [mean_left_dwell],
        'std_left_dwell': [std_left_dwell],
        'min_mid_dwell': [min_mid_dwell],
        'max_mid_dwell': [max_mid_dwell],
        'mean_mid_dwell': [mean_mid_dwell],
        'std_mid_dwell': [std_mid_dwell],
        'min_right_dwell': [min_right_dwell],
        'max_right_dwell': [max_right_dwell],
        'mean_right_dwell': [mean_right_dwell],
        'std_right_dwell': [std_right_dwell]
    })], ignore_index=True)

# Print the results
print(results)
```

Statistical Tests

```python
import pandas as pd
from scipy import stats

# Assuming you have your DataFrame 'sample_df' with features and 'label' column

# Select the features (independent variables) and the target (dependent variable)
X_label0 = sample_df[sample_df['label'] == 0].drop(['label', 'gene_id', 'transcript_id', 'k_mer'], axis=1)
X_label1 = sample_df[sample_df['label'] == 1].drop(['label', 'gene_id', 'transcript_id', 'k_mer'], axis=1)

# Perform a t-test for each feature
for feature in X_label0.columns:
    t_stat, p_value = stats.ttest_ind(X_label0[feature], X_label1[feature], equal_var=False)
    print(f'Test for feature: {feature}')
    print(f'T-statistic: {t_stat}')
    print(f'P-value: {p_value}')

    # You can set a significance level (alpha) and check if the p-value is below it
    alpha = 0.05
    if p_value < alpha:
        print(f'Statistically significant difference for {feature}')
    else:
        print(f'No statistically significant difference for {feature}')
    print()
```

Observation:

- **Transcript Position, Left Dwell, Left Std, Left Mean, Mid Dwell, Right Dwell, Right Std, Right Mean, D1, D2, R, H1, and H2**: The p-values for these features are essentially zero, and the t-statistics are quite large in magnitude, which strongly indicates that there are statistically significant differences for these features between the groups being compared. This significance is robust as the p-values are far below the common alpha level of 0.05, often used as a threshold for significance.

- **Mid Std**: This is the only feature where the p-value (0.1788) is above the typical alpha level of 0.05, and the t-statistic is relatively small. This indicates that there is no statistically significant difference for 'mid_std' between the groups being tested.

Interpretation:
- **Highly Significant Features**: For features with p-values at or near zero, the negative t-statistics for features like 'transcript_position', 'left_dwell', 'left_std', 'left_mean', 'mid_dwell', 'right_dwell', 'right_std', and 'right_mean' suggest that the mean of these features is significantly lower in one group compared to the other. Conversely, the positive t-statistics for 'D1', 'R', 'H1', and 'H2' suggest their means are significantly higher in one group compared to the other.

- **Non-significant Feature**: The feature 'mid_std' does not show a significant difference between groups, indicating that the variation in 'mid' dwell times is similar across the groups.

- **Magnitude of T-Statistics**: The magnitude of the t-statistic indicates the degree of difference between the groups. Extremely high absolute values, as seen with 'left_mean', 'mid_mean', 'D2', 'R', 'H1', and 'H2', suggest very pronounced differences. This is consistent with the feature importance observed from the Decision Tree Classifier, where 'transcript_position' was shown as the most important feature.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller

# Create a synthetic time series using 'transcript_position' as time
data = {
    'gene_id': [5811021, 888795, 230794, 5912941, 10111111],
    'transcript_position': [944, 306, 1638, 43, 952],
    'left_dwell': [0.013, 0.006, 0.004, 0.004, 0.019],
    'mid_dwell': [0.005, 0.003, 0.006, 0.012, 0.021],
    'right_dwell': [0.007, 0.009, 0.008, 0.005, 0.003]
}

# Create a DataFrame from the synthetic data
df = pd.DataFrame(data)

# Set 'transcript_position' as the index and sort by it (simulating a time series)
df.set_index('transcript_position', inplace=True)
df.sort_index(inplace=True)

# Decompose the synthetic time series
decomposition = seasonal_decompose(df['left_dwell'], model='additive')
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plot the decomposed components
plt.figure(figsize=(12, 6))
plt.subplot(411)
plt.plot(df['left_dwell'], label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal, label='Seasonal')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residual')
plt.legend(loc='best')

# Perform Augmented Dickey-Fuller test for stationarity
result = adfuller(df['left_dwell'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:', result[4])

# Further analysis, modeling, and forecasting can be done based on the decomposition and stationarity test results.
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt

# Create a synthetic time series (as shown in the previous example)
# ...

# Prepare lag features (use the 'left_dwell' column as the target variable)
df['left_dwell(t-1)'] = df['left_dwell'].shift(1)
df['left_dwell(t-2)'] = df['left_dwell'].shift(2)

# Drop missing values
df.dropna(inplace=True)

# Split the data into training and testing sets
train_size = int(0.8 * len(df))
train_data, test_data = df[:train_size], df[train_size:]

# Define features and target variable
X_train = train_data[['left_dwell(t-1)', 'left_dwell(t-2)']]
y_train = train_data['left_dwell']
X_test = test_data[['left_dwell(t-1)', 'left_dwell(t-2)']]
y_test = test_data['left_dwell']

# Create and train a Random Forest Regressor
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf_regressor.predict(X_test)

# Plot the original and predicted time series
plt.figure(figsize=(12, 6))
plt.plot(test_data.index, y_test, label='Original', marker='o')
plt.plot(test_data.index, y_pred, label='Predicted', marker='x')
plt.legend(loc='best')
plt.show()
```
